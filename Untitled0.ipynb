```python
import cv2
import numpy as np
import requests
import os
import shutil
from google.colab import drive
from google.colab.patches import cv2_imshow

# 1. 连接云盘（为了保存最终的“高清成品”）
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

# 2. 从网上下载一张“素材图”到临时磁盘 /content/
# 这张图约 2MB，关掉网页它就消失，不占你云盘空间
img_url = "https://raw.githubusercontent.com/opencv/opencv/master/samples/data/fruits.jpg"
input_path = "/content/raw_material.jpg"

r = requests.get(img_url)
with open(input_path, 'wb') as f:
    f.write(r.content)

# 3. 开始“加工”：AI 锐化算法
img = cv2.imread(input_path)

# 创建一个锐化算子（让模糊的边缘变清晰）
kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
sharpened = cv2.filter2D(img, -1, kernel)

# 保存处理后的结果到临时空间
result_path = "/content/refined_result.jpg"
cv2.imwrite(result_path, sharpened)

# 4. 预览对比（左边是原图，右边是修复图）
print("对比预览：左侧为原图，右侧为锐化后的效果")
combined = np.hstack((img, sharpened)) # 把两张图拼在一起看
cv2_imshow(combined)

# 5. 【关键动作】只把 200KB 的成品搬进云盘，抛弃 2MB 的素材
destination_folder = '/content/drive/MyDrive/My_AI_Work/'
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

shutil.move(result_path, destination_folder + 'high_definition_fruits.jpg')

print(f"\n✅ 处理完成！成品已存入云盘：{destination_folder}")
print("🗑️  原始素材留在临时磁盘，即将随网页关闭而自动清理。")
```
